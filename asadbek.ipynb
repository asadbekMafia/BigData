```python
import matplotlib.pyplot as plt
import pandas as pd
from google.colab import files

# 1️⃣ Faylni yuklash
uploaded = files.upload()
filename = list(uploaded.keys())[0]

# 2️⃣ CSV faylni o'qish (UTF-8 BOM bilan)
df = pd.read_csv(filename, sep=';', encoding='utf-8-sig')

# 3️⃣ Ustunlarni ko'rish
print("Fayldagi ustunlar:", df.columns)

# 4️⃣ Indexni x-o‘qda ishlatish va ikkita ustunni grafik qilish
plt.figure(figsize=(12,6))
plt.plot(df.index, df['ACE'], marker='o', label='ACE', color='blue')
plt.plot(df.index, df['System imbalance'], marker='x', label='System imbalance', color='red')
plt.title('ACE va System imbalance taqqoslanishi')
plt.xlabel('Index')
plt.ylabel('Qiymat')
plt.legend()
plt.grid(True)
plt.show()

# 5️⃣ Korrelatsiya hisoblash
correlation = df['ACE'].corr(df['System imbalance'])
print(f"ACE va System imbalance orasidagi korrelatsiya: {correlation:.2f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import xml.etree.ElementTree as ET
from google.colab import files

# 1️⃣ Faylni yuklash
uploaded = files.upload()
filename = list(uploaded.keys())[0]

# 2️⃣ CSV faylni o'qish (UTF-8 BOM bilan)
df = pd.read_csv(filename, sep=';', encoding='utf-8-sig')

# 3️⃣ Ustunlarni ko'rish
print("Fayldagi ustunlar:", df.columns)



# 5️⃣ XML formatda saqlash
root = ET.Element('Data')
for idx, row in df.iterrows():
    entry = ET.SubElement(root, 'Entry')
    ET.SubElement(entry, 'Index').text = str(idx)
    ET.SubElement(entry, 'ACE').text = str(row['ACE'])
    ET.SubElement(entry, 'System_imbalance').text = str(row['System imbalance'])

tree = ET.ElementTree(root)
tree.write('data.xml')

# 6️⃣ JSON formatda saqlash
df.to_json('data.json', orient='records', indent=4)

# 7️⃣ Excel formatda saqlash
df.to_excel('data.xlsx', index=False)

# 8️⃣ Fayllarni Colab’dan yuklab olish
files.download('data.xml')
files.download('data.json')
files.download('data.xlsx')

print("Ma'lumotlar XML, JSON va Excel formatida saqlandi va yuklashga tayyor.")
```
